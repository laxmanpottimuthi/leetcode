```python
"""
Part 2: Write a small, reasonably efficient class that takes strings as input, and calls
a callback method whenever a newline ('\n') is reached, passing in the previous line's
text.  The class should discard old data when possible to save memory.

Hint: This should not take more than about 15-50 lines of code.
"""
class LineParser:
    def __init__(self, callback):
        self.callback = callback
        self.buffer = ""

    def read(self, data):
        # TODO Fill in this funtion
        for i in data:
            if i != "\n":
                self.buffer += i
            else:
                self.callback(self.buffer)
                self.buffer = ""

parser = LineParser(print)
parser.read("This is t")
parser.read("he first l")
parser.read("ine.\nAnd this is the second.\n")
parser.read("And this is the third.\nAnd the")
parser.read(" fourth.\n")
```

```python
"""
Part 3: Implement Ultimate Tic Tac Toe

Tic-Tac-Toe is a game played between player X and player O on a 3x3 grid.

Starting with X, players alternate placing their mark in an empty square. If a player has three consecutive
marks orthogonally or diagonally, he wins.  If there is no available move, the game is tied.

   X | O | O
  -----------
   O | X |      (X wins)
  -----------
   X |   | X

*Ultimate* Tic-Tac-Toe is played on 3x3 grid of Tic-Tac-Toe games. The objective is to win 'local' games
that give you three wins in-a-row on the 'global' game.

  _|_|_  _|_|_  _|_|_
  _|_|_  _|_|_  _|_|_
   | |    | |    | |

  _|_|_  _|_|_  _|_|_
  _|_|_  _|_|_  _|_|_
   | |    | |    | |

  _|_|_  _|_|_  _|_|_
  _|_|_  _|_|_  _|_|_
   | |    | |    | |

The game starts with X playing wherever they want in any of the 81 empty spots. This move 'sends' their
opponent to its relative location. For example, if 'X' played in the top right square of their local board,
then 'O' needs to play next in the local board at the top right of the global board. 'O' can then play in
any one of the nine available spots in that local board, each move sending 'X' to a different local board.

Once the outcome of a local board is decided (won or the board is full), no more moves may be played in
that board. If a player is sent to such a board, then that player may play in any other board.

The first player to win three local games orthogonally or diagonally on the global board wins Ultimate
Tic-Tac-Toe. If there is no available move, the game is tied.

Game rules:
https://en.wikipedia.org/wiki/Ultimate_tic-tac-toe
Online game with valid-move highlighting:
http://ultimatetictactoe.creativitygames.net/  (Change "First Tile Wins" to "3 Tiles in a Row")

Extra Credit:
- Implement a computer player that always makes a valid move
- Implement a computer player that makes strategic move
"""
def part_3():
    PIECE_X = 'X'
    PIECE_O = 'O'
    PIECE_EMPTY = '_'
    COMP = False

    class Coordinate:
        def __init__(self, x, y):
            self.x = x
            self.y = y

    class Game:
        def __init__(self):
            self.size = 9
            self.board = [
                [PIECE_EMPTY] * self.size
                for r in range(self.size)
            ]
            self.local_board_status = [('_',False) for i in range(9)]
            self.PIECE_X = 'X'
            self.PIECE_O = 'O'
            self.PIECE_EMPTY = '_'
            self.next_board = None
            self.current_player = self.PIECE_X
            self.prev_player = None
            self.board_change = False

        def get_user_move(self,player):
            try:
                print("Player ",player)
                user = input("Move: ")
                assert(len(user) == 2)
                move = Coordinate(
                    ord(user[0]) - ord('a'),
                    int(user[1]) - 1
                )
                assert(move.x > -1 and move.x < 9)
                assert(move.y > -1 and move.y < 9)
                print()
                return move
            except (KeyboardInterrupt, EOFError) as e:
                exit()
            except:
                print("Invalid input. Format should be like b4\n")
                return self.get_user_move(player)
        
       
        def get_local_board(self,move):
            return move.x//3+move.y//3
            
        def set_local_board_status(self,local_board,player):
            self.local_board_status[local_board] = (player,True)
        
        def check_local_board_status(self,local_board_number):
            local_board = local_board_number
            #position in global board
            x = local_board_number//3
            y = local_board_number%3
            x *= 3
            y *= 3
            
            for i in range(x,x+3):
                count_x,count_o = 0,0
                for j in range(y,y+3):
                    if self.board[i][j] == self.PIECE_X:
                        count_x += 1
                    if self.board[i][j] == self.PIECE_O:
                        count_o += 1
            if count_x == 3:
                return True,self.PIECE_X
            if count_o == 3:
                return True,self.PIECE_O
            
            x,y = y,x
            for i in range(x,x+3):
                count_x,count_o = 0,0
                for j in range(y,y+3):
                    if self.board[i][j] == self.PIECE_X:
                        count_x += 1
                    if self.board[i][j] == self.PIECE_O:
                        count_o += 1
            if count_x == 3:
                return True,self.PIECE_X
            if count_o == 3:
                return True,self.PIECE_O
            
            count_x,count_o = 0,0
            for i in range(0,3):
                if self.board[x+i][y+j] == self.PIECE_X:
                        count_x += 1
                if self.board[x+i][x+j] == self.PIECE_O:
                    count_o += 1
            if count_x == 3:
                return True,self.PIECE_X
            if count_o == 3:
                return True,self.PIECE_O
            
            return False,self.PIECE_EMPTY
        def check_global_board_status(self,local_board_number):
            x_r = local_board_number//3
            y_c = local_board_number%3
            x = x_r*3
            y = y_c*3
                
            count_x,count_o = 0,0
            for i in range(x,x+3):
                player, board_status = self.local_board_status[i]
                
                if board_status:
                    if player == self.PIECE_X:
                        count_x += 1
                    if player == self.PIECE_O:
                        count_o += 1
                if count_x == 3 or count_o == 3:
                    won_player = self.PIECE_X if count_x>count_o else self.PIECE_O
                    print("Player: ", won_player, " won the game")
                    return True, won_player
            count_x,count_o = 0,0
            for i in [y_c, y_c+3, y_c+6]:
                player, board_status = self.local_board_status[i]
                
                if board_status:
                    if player == self.PIECE_X:
                        count_x += 1
                    if player == self.PIECE_O:
                        count_o += 1
                if count_x == 3 or count_o == 3:
                    won_player = self.PIECE_X if count_x>count_o else self.PIECE_O
                    print("Player: ", won_player, " won the game")
                    return True, won_player
            
            if local_board_number % 2 == 0:
                count_x,count_o = 0,0
                for i in [0,4,8]:
                    player, board_status = self.local_board_status[i]
                
                    if board_status:
                        if player == self.PIECE_X:
                            count_x += 1
                        if player == self.PIECE_O:
                            count_o += 1
                    if count_x == 3 or count_o == 3:
                        won_player = self.PIECE_X if count_x>count_o else self.PIECE_O
                        print("Player: ", won_player, " won the game")
                        return True, won_player
                for i in [2,4,6]:
                    player, board_status = self.local_board_status[i]
                
                    if board_status:
                        if player == self.PIECE_X:
                            count_x += 1
                        if player == self.PIECE_O:
                            count_o += 1
                    if count_x == 3 or count_o == 3:
                        won_player = self.PIECE_X if count_x>count_o else self.PIECE_O
                        print("Player: ", won_player, " won the game")
                        return True, won_player
                
            return False,'_'
        def check_for_board_validity(self,board):
            if self.next_board :
                return True
            player, board_status = self.local_board_status[board]
            if board_status:
                print("Local board already filled by player ", player)
            
            
        # Things to remember:
        # 1. Take turns
        # 2. Valid moves (send opponent to correct local board)
        # 3. Detect when sub-section is won/full (don't allow for futher play in section)
        # 4. Detect when global game is won/tied (declare winner and terminate)
        def play(self):
            # TODO Implement game logic
            
            while True:
                self.render()
                if not COMP:
                    move = self.get_user_move(self.current_player)
                else:
                    print("Not implemented")
                
                
                
                local_board_number = self.get_local_board(move)
                
                if not self.check_for_board_validity(local_board_number) or self.next_board != local_board_number:
                    print("Invalid move.")
                    continue
                
                self.prevBoard = local_board_number 
                
                self.board[move.x][move.y] = self.current_player
                local_board_status,player = self.check_local_board_status(local_board_number)
                if local_board_status:
                    self.set_local_board_status(local_board_number,player)
                    global_board_status, player_won = self.check_global_board_status(local_board_number)
                    if global_board_status:
                        self.render()
                        if player_won != self.current_player:
                            print("Something went wrong")
                        print("Player ",self.current_player, " won the game")
                        break
                if self.current_player == self.PIECE_X:
                    self.current_player = self.PIECE_O 
                    self.prev_player = self.PIECE_X
                else:
                    self.current_player = self.PIECE_X
                    self.prev_player = self.PIECE_O
                self.prevBoard = local_board_number

        def render(self):
            print("  a b c   d e f   g h i")
            print("\n  ______|_______|______\n        |       |\n".join(
                    map(lambda outer_y: "\n".join(
                            map(lambda inner_y: str(outer_y * 3 + inner_y + 1) + " " + " | ".join(
                                    map(lambda outer_x: "|".join(
                                            map(
                                                lambda inner_x: self.board[outer_x * 3 + inner_x][outer_y * 3 + inner_y],
                                                range(3)
                                            )
                                        ), range(3)
                                    )
                                ), range(3)
                            )
                        ), range(3)
                    )
                )
            )
            print()

    return Game

game = part_3()()
game.play()
```
