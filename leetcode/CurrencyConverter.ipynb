```python
def isCurrency(strAmount):
    euro = u"\N{euro sign}"
    yen = u"\N{yen sign}"
    
    currencies = set([euro,yen,'$'])
    digits = set(str(i) for i in range(10))
    sign = set(["-"])
    brackets = set(["(",")"])
    import operator
    from functools import reduce
    validChar = reduce(operator.or_, [ currencies,digits,sign,brackets,set([".",","])])
    lenStr = len(strAmount)
    decimal = 0
    coma = currencyCnt = signCt = 0
    parantheses = 0
    prev = None
    digitCt = 0
    for i in range(lenStr-1, -1, -1):
        char = strAmount[i]
        if char not in validChar or currencyCnt > 1 or char == " ":
            return False
        if char in currencies:
            if char == yen:
                return False
            currencyCnt += 1
            if currencyCnt != 1:
                return False
            
        if char ==")":
            parantheses += 1
            if parantheses > 1 or digitCt > 0:
                return False
        if char == "(":
            if signCt:
                return False
            prev = "("
            parantheses -= 1
            if parantheses:
                return False
        
        if char in sign:
            if prev in brackets or i != 0:
                return False
            
        if char == ".":
            if digitCt != 2:
                return False
            digitCt = 0
            decimal += 1
            if decimal != 1:
                return False
        if char == ",":
            if digitCt != 3:
                return False
            digitCt = 0
            
        if char in digits:
            digitCt += 1
            
    return True
```
